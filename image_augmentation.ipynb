```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from augmentation import drawbox, Patch
```

```python
imgpath = './test.jpg'
img = cv2.imread(imgpath)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

bbox = (152,210,445,260)
```

```python
p1 = Patch(img,bbox)
p1.show()
```

```python
p1.pad = (150,20)
for mag in [(0.5,0.5),(0.5,2),(1,1),(2,0.5),(2,2)]:
    fig, ax = plt.subplots(nrows=2,ncols=4)
    fig.set_figheight(10)
    fig.set_figwidth(20)
    for i,flp in enumerate([True,False]):
        if flp:
            flp_str = 'flip'
        else:
            flp_str = 'not flip'
        for j,rot in enumerate([0,90,180,270]):
            p1.mag = mag
            p1.flip = flp
            p1.rotation = rot
            ax[i,j].set_title(f'{flp_str}, rotation : {rot}')
            p1.show(ax[i,j])
    plt.show()
        
p1.mag=(0.5,0.5)
p1.show()
```

```python
p1.pad=0,0
defaultmag = 0.5
for mag in [(0.5,0.5),(0.5,1),(1,1),(1,0.5)]:
    fig, ax = plt.subplots(nrows=2,ncols=4)
    fig.set_figheight(10)
    fig.set_figwidth(20)
    for i,flp in enumerate([True,False]):
        if flp:
            flp_str = 'flip'
        else:
            flp_str = 'not flip'
        for j,rot in enumerate([0,90,180,270]):
            p1.mag = tuple(x*defaultmag for x in mag)
            p1.flip = flp
            p1.rotation = rot
            ax[i,j].set_title(f'{flp_str}, rotation : {rot}')
            imover,bbox = p1.overlay(img,100,100)
            ax[i,j].imshow(imover)
            drawbox(*bbox,ax=ax[i,j])
            
#             p1.show(ax[i,j])
    plt.show()
```
